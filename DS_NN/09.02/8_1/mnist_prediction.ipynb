```python
from keras.models import load_model

import matplotlib.pyplot as plt

import numpy
```

```python
img = plt.imread("9.png")
img = img[:, :, 1]
```

```python
plt.imshow(img)
```

```python
img = numpy.array(img.reshape(28, 28, 1))

img = 1 - img

img = numpy.array([img])

img
```

```python
loaded_model = load_model('mnist_base_10.keras')
```

```python
prediction = loaded_model.predict(img)
```
